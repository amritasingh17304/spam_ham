```
data=pd.read_csv("/content/drive/My Drive/spam ham data set.csv",encoding="Latin-1")
data.head()
```

```
data=data.drop_duplicates()
data.head()
```

```
for i in string.punctuation:
  data['v2']=data['v2'].str.replace(i,'')
data['v2']=data['v2'].str.lower()
data.head()
```

```
tr=0.75*data.shape[0]
train=data.iloc[range(0,int(tr))]
test=data.iloc[range(int(tr),data.shape[0])]
```

```
train['v1'].value_counts()
```

```
pspam=536/train.shape[0]
pham=1-pspam
```

```
words_spam=defaultdict(int)
for i in train.index:
  if(train['v1'][i]=='spam'):
    for j in train['v2'][i].split():
      words_spam[j]+=1
# print(words_spam)
spamdict=defaultdict(float)
k=np.sum(list(words_spam.values()))
# print(words_spam.values())
for i in words_spam.keys():
  spamdict[i]=(words_spam[i])/(536)
print(spamdict)

  
```

```
words_ham=defaultdict(int)
for i in train.index:
  if(train['v1'][i]=='ham'):
    for j in train['v2'][i].split():
      words_ham[j]+=1
#print(words_ham)
hamdict=defaultdict(float)
k=np.sum(list(words_ham.values()))
for i in words_ham.keys():
  hamdict[i]=(words_ham[i])/(3599)
print(hamdict)
```

```
words=defaultdict(int)
for i in train.index:
  for j in train['v2'][i].split():
    words[j]+=1
print(words)
```

```
k=np.sum(list(words.values()))
wordsdict=defaultdict(float)
for i in words.keys():
  wordsdict[i]=words[i]/k
print(wordsdict)
  
```

```
for i in wordsdict.keys():
  wordsdict[i]=np.log(1+wordsdict[i])
for i in spamdict.keys():
  spamdict[i]=np.log(1+spamdict[i])
for i in hamdict.keys():
  hamdict[i]=np.log(1+hamdict[i])

```

```
s=0
x=0
fail_s=0
fail_x=0
for i in test.index:
  spam_prob=np.log(pspam)
  ham_prob=np.log(pham)
  for j in test['v2'][i].split():
    spam_prob+=spamdict[j]
    ham_prob+=hamdict[j]
  if(spam_prob>ham_prob):
    if(test['v1'][i]=='spam'):
      s+=1
    else:
      fail_s+=1
  else:
    if(test['v1'][i]=='ham'):
      x+=1
    else:
      fail_x+=1
print(s)
print(fail_s)
print(x)
print(fail_x)
print((s+x)/(s+x+fail_s+fail_x))

```
